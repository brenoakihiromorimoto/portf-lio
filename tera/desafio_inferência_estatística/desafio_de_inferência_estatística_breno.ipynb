### 5 - Testes de hipóteses

1. Faça o(s) teste(s) de hipóteses adequado(s) para avaliar a significância estatística das diferenças nas características demográficas apontadas na análise bivariada. Quais são as características apresentam diferenças estatisticamente significativas para a frequência de sintomas de depressão?
* Renda familiar: stat=-0.440, p=0.660
    * Hipótese nula
* Gênero: stat=-2.175, p=0.030
    * Hipótese alternativa
* Idade: stat=-1.625, p=0.104
    * Hipótese nula
* Etnia: stat=-0.758, p=0.449
    * Hipótese nula
* Escolaridade: stat=4.007, p=0.000
    * Hipótese alternativa
* Gênero e escolaridade são características estatisticamente significativas de acordo com o p-valor para sintomas de depressão.

Teste-t independente

* Hipótese nula: média de renda familiar do grupo sem sintomas == média de renda familiar do grupo com sintomas severos
* Hipótese alternativa: média de renda familiar do grupo sem sintomas != média de renda familiar do grupo com sintomas severos 

```python
from scipy.stats import ttest_ind
stat, p = ttest_ind(phq_full[phq_full['phq_grp']=='sem sintomas']['Renda_familiar'], phq_full[phq_full['phq_grp']=='sintomas severos']['Renda_familiar'])
print('Renda familiar: stat=%.3f, p=%.3f' % (stat, p))
if p>0.05:
    print('Hipótese nula')
else:
     print('Hipótese alternativa')
stat, p = ttest_ind(phq_full[phq_full['phq_grp']=='sem sintomas']['Gênero'], phq_full[phq_full['phq_grp']=='sintomas severos']['Gênero'])
print('Gênero: stat=%.3f, p=%.3f' % (stat, p))
if p>0.05:
    print('Hipótese nula')
else:
     print('Hipótese alternativa')
stat, p = ttest_ind(phq_full[phq_full['phq_grp']=='sem sintomas']['Cat_idade_int'], phq_full[phq_full['phq_grp']=='sintomas severos']['Cat_idade_int'])
print('Idade: stat=%.3f, p=%.3f' % (stat, p))
if p>0.05:
    print('Hipótese nula')
else:
     print('Hipótese alternativa')

stat, p = ttest_ind(phq_full[phq_full['phq_grp']=='sem sintomas']['Etnia'], phq_full[phq_full['phq_grp']=='sintomas severos']['Etnia'])
print('Etnia: stat=%.3f, p=%.3f' % (stat, p))
if p>0.05:
    print('Hipótese nula')
else:
     print('Hipótese alternativa')
        
stat, p = ttest_ind(phq_full[phq_full['phq_grp']=='sem sintomas']['Escolaridade'], phq_full[phq_full['phq_grp']=='sintomas severos']['Escolaridade'])
print('Escolaridade: stat=%.3f, p=%.3f' % (stat, p))
if p>0.05:
    print('Hipótese nula')
else:
     print('Hipótese alternativa')
```

 ### 6 - Aspectos da análise

1. Qual o tipo de estudo está sendo empregado pelo NHANES? experimental ou observacional?
* Observacional, os pesquisadores não podem forçar as pessoas do estudo a ingerirem certos alimentos e fazerem uma quantidade de exercício físico por semana.
* Vantagem:
* Alta validade externa (cenário mais próximo da vida real)
* Desvantagens:
* Em geral, não permitem interpretação causal
* A interpretação causal requer suposições fortes

2. Discuta as possíveis fontes de viés presentes na análise.
* Viés de confundimento: o efeito da variável explicativa se mistura aos efeitos de variáveis confundidoras.
    * Gênero feminino está associado aos maiores sintomas de depressão?
        * Variável confundidora: pode ser TPM, depressão pós-parto ou dupla jornada de trabalho.
    * Baixa escolaridade está associada aos maiores sintomas de depressão?
        * Variável confundidora: problemas familiares, bullying, violência doméstica, pobreza extrema, envolvimento com atividades ilícitas, entre outras.
    * A ingestão de laticínios está associada aos maiores sintomas de depressão?
        * Variável confundidora: intolerância à lactose.
    * A ingestão de açúcares está associada aos maiores sintomas de depressão?
        * Variável confundidora: falta de sono, estresse na vida pessoal ou profissional.
    * A menor quantidade de minutos que alguém faz exercícios físicos por semana está associada aos maiores sintomas de depressão?
        * Variável confundidora: estresse na vida pessoal ou profissional, falta de sono, entre outros.

4. Quais são as fraquezas das análises realizadas? O que pode ser feito para melhorar?
* As fraquezas são os dados insuficientes para estabelecer uma relação causal entre hábitos saudáveis e sintomas de depressão.
* Poucas inferências podem ser feitas a partir de uma grande quantidade de dados.
* Muitas varáveis confundidoras.
* O que pode ser melhorado:
* 1. Controlar (incluir na análise) variáveis confundidoras conhecidas. Iremos supor que o efeito de confundimento de variáveis não coletadas é mínimo
* 2. Matching: selecionar o grupo controle de acordo com características do grupo de exposição (ex.: idade, gênero, localização, etc.)
